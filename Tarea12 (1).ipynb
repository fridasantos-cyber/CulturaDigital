# Clase: Booleanos y Ciclo `while` en Python

## ¿Qué es un valor booleano?
Los **booleanos** son un tipo de dato que representan valores de verdad, es decir si algo es verdadero o falso. Cundo usamos en el `if` la condicion tomaba un valor de verdad y dependiendo de su valor realizabamos alguna acción; es decir, los usamos para tomar decisiones en los programas.. En python tendremos los valores `True` para el valor de verdadero y `False` para el valor de falso.


```python
a = True
b = False
print("a:", a)
print("b:", b)
```

Nota que pasa si mandamos a pantalla condiciones parecidas a las que usamos en los ejercicios de las condicionales.

```python
print("Ejemplo uno: ",5<6)
print("Ejemplo dos: ",5>6)
print("Ejemplo tres: ",5==6)
```

## **Ciclo `while`**



El ciclo `while` repite un bloque **mientras** una condición sea verdadera.

```python
#Ejemplo:

#Definimos una variable llamada contador con el valor de 1.
contador = 1

#Mientras la variable tenga un valor menor a 5 vamos a repetir el código siguiente.
while contador < 5:
    #Mandamos a pantalla Contador seguido del valor que tiene la varibale contador.
    print("Contador:", contador)
    #Aumentamos el valor de contador
    contador = contador + 1
```

Nota que en el código pasado dentro del while la condición que se está evaluando es si el valor de la variable `contador` es menor a 5; y justo ese es el valor booleano, mientras sea cierto el valor de `contador < 5` es `True`; en el momento en que no se cumpla el valor de `contador < 5` será `False`.



Veamos otro ejmeplo, pero ahora le pasaremos explicitamente un valor boolenao a la condición del while.


```python
#Definimos una variable llamada activo con el valor boolenao True
activo = True
#Definimos una variable llamada número con el valor 0
numero = 0

#Mientras la variable activo tenga el valor de True realiza el siguiente código.
while activo:
    #Manda a pantalla Numero seguido del valor de la variable número.
    print("Número:", numero)
    #Aumenta el valor de la variable número en uno.
    numero = numero + 1
    #Si el valor de la variable numero es igual a 4 realiza lo siguiente.
    if numero == 4:
        #Cambia el valor de la varibale activo a False.
        activo = False
```

1. Usa un ciclo `while` para imprimir los números del 1 al 10.

```python
#Realiza el ejercicio 1.
# Inicializo una variable llamada num con el valor 1
num = 1

# Inicio un ciclo while que seguirá mientras num sea menor o igual a 10
while num <= 10:
    print(num)  # Imprime el valor actual de num
    num += 1     # Incrementa num en 1 para avanzar al siguiente número
```

2. Crea una variable booleana llamada `sigue` y úsala para controlar un ciclo que imprima "Hola" 5 veces.


```python
#Realiza el ejercicio 2.
# Creo la variable booleana 'sigue' que controlará el ciclo
sigue = True

# Creo un contador para llevar la cuenta de cuántas veces se imprime "Hola"
contador = 0

# El ciclo se repetirá mientras 'sigue' sea True
while sigue:
    print("Hola")    # Imprime la palabra Hola
    contador += 1     # Aumento el contador en 1

    # Cuando el contador llegue a 5, cambiamos sigue a False para detener el ciclo
    if contador == 5:
        sigue = False
```

3. Escribe un `while` que imprima solo números pares hasta llegar al 20.


```python
#Realiza el ejercicio 3.
# Inicializo una variable llamada num con el valor 2, ya que es el primer número par
num = 2

# El ciclo continúa mientras num sea menor o igual a 20
while num <= 20:
    print(num)    # Imprime el número par actual
    num += 2       # Incrementa num en 2 para obtener el siguiente número par
```

4. Crea un programa que pida al usuario una contraseña usando `input()` repita mientras la contraseña sea incorrecta. La contraseña correcta es "python123".


```python
#Realiza el ejercicio 4.
# Solicito al usuario que ingrese una contraseña por primera vez
password = input("Ingresa la contraseña: ")

# Mientras la contraseña ingresada no sea la correcta, seguirá pidiendo otra
while password != "python123":
    print("Contraseña incorrecta, intenta de nuevo.")
    password = input("Ingresa la contraseña: ")  # Vuelve a pedir contraseña

# Cuando la contraseña sea correcta sale del ciclo
print("¡Contraseña correcta! Acceso concedido.")
```
