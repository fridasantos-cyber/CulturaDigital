### ***Ejercicio 1:***  
Pide dos números al usuario y muestre en pantalla el resultado de multiplicarlos.

````python
#Respuestas Ejercicio1
numero1 = float(input("Introduce el primer número: "))
numero2 = float(input("Introduce el segundo número: "))
resultado = numero1 * numero2
print("El resultado de multiplicar", numero1, "por", numero2, "es:", resultado)

````

### ***Ejercicio 4:***
Pide al usuario el precio de un producto y la cantidad que va a comprar, luego muestra el total a pagar.

````python
#Respuesta de ejercicio 4
precio = float(input("Ingresa el precio del producto: "))
cantidad = int(input("Ingresa la cantidad que vas a comprar: "))

total = precio * cantidad

print(f"El total a pagar es: {total}")
````

### ***Ejercicio 5:***
Pide al usuario una cantidad de minutos y muestra cuántos segundos equivalen.

````python
#Respuesta de ejercicio 5
minutos = float(input("Ingresa la cantidad de minutos: "))

segundos = minutos * 60

print(f"{minutos} minutos equivalen a {segundos} segundos.")
````
